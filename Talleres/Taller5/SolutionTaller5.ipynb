```python
%pylab inline 
from scipy.optimize import basinhopping
```

```python

def Oncuadrada(x,Amplitud,T,numTerm):
    suma=0
    for i in range(1,1+numTerm):
        suma+=(Amplitud)*(4/pi)*sin(2*pi*(2*i-1)*(1./T)*x)*(1./(2.*i-1))
    return suma

y=[]
xlin=linspace(0,2*pi,100)
for i in range(len(xlin)):
     y.append(Oncuadrada(xlin[i],pi/4,2*pi,10))


figure()
plot(xlin,y)
show()
```

```python
def cteGibbs (Oncuadrada,numTerm):
    maxi=basinhopping(lambda x: Oncuadrada(x,pi/4.,2.*pi,numTerm),2,niter=100)
    return (maxi.fun*-2)

numTerm=range(5,50,5)
consG=[]
for i in numTerm:
    consG.append((cteGibbs(Oncuadrada,i)-pi/2.)/pi)

figure(figsize(7,5))
ylim(0.0890,0.092)
scatter(numTerm,consG)
title("Constante de Gibbs")
xlabel("número de términos")
ylabel("$(G-\pi/2)/\pi$")
show()


    
```
