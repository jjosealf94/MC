```python
%pylab inline
from scipy.optimize import basinhopping
```

```python
def Oncuadrada(x,Amplitud,T,numTerm):
    suma=0
    for i in range(1,1+numTerm):
        suma+=(Amplitud)*(4/pi)*sin(2*pi*(2*i-1)*(1./T)*x)*(1./(2.*i-1))
    return suma

y=[]
xlin=linspace(0,2*pi,100)

for i in range(len(xlin)):
     y.append(Oncuadrada(xlin[i],pi/4,2*pi,10))

figure()
plot(xlin,y)
show()
```

```python
def maxi(Oncuadrada,N):
    mini= basinhopping(lambda x: Oncuadrada(x,pi/4,2*pi,N),1,niter=200)
    return abs(mini.fun*(2))

N=range(1,30)
cteGibbs=[]
for i in range(len(N)):
    cteGibbs.append((maxi(Oncuadrada,N[i])-pi/2)/pi)
scatter(N,cteGibbs)
ylabel("$(G-\pi\/2)/\pi$")
xlabel("N")
title("Constante de Gibbs")
show()

```
