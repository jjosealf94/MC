```python
%pylab inline 
from scipy.integrate import odeint
from scipy.integrate import ode
from mpl_toolkits.mplot3d import Axes3D
```

```python
a=0.1
b=0.1
c=14
# Las ecuaciones del sistema de RÃ¶ssler X 10
def Rosseler(Y,t):
    return np.array([(-Y[1]-Y[2]),Y[0]+(a*Y[1]),b+(Y[2]*(Y[0]-c))]) 
```

```python
tmax=600
times=np.linspace(0,tmax,10000)
funsol,others=odeint(Rosseler,[1.,1.,1.],times,full_output=True,printmessg=True)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')
plt.title("Atractor de RÃ¶ssler x 10")
plt.plot(funsol[:,0],funsol[:,1],funsol[:,2])
plt.show()
```
