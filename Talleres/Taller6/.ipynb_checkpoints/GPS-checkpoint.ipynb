```python
%pylab inline 
import numpy as np
```

```python
data = genfromtxt("./running_clean_nov.csv", delimiter=',')
lon=data[:,0]
lat=data[:,1]
hour=data[:,3]
minu=data[:,4]
seg=data[:,5]
```

```python
r= 6400
x= r*np.sin(90-lat)*np.cos(lon)
y=r*np.cos(90-lat)*np.cos(lon)


uniOriente=[-np.sin(-74.0758),np.cos(-74.0758),0]
uniNorte=[np.cos(4.5981),np.sin(4.5981),0]

yBog=np.zeros(len(y))
xBog=np.zeros(len(x))

for i in range(len(x)):
    xBog[i]=x[i]*uniOriente[0]+y[i]*uniOriente[1]
    yBog[i]=x[i]*uniNorte[0]+y[i]*uniNorte[1]


desxBog=np.zeros(len(xBog))
desyBog=np.zeros(len(yBog))

for i in range(len(xBog)):
    desxBog[i]=xBog[i]-xBog[0]
    desyBog[i]=yBog[i]-yBog[0]

figure(figsize(6,6))
plot(desxBog,desyBog)
show()
```
