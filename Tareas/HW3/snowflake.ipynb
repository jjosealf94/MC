```python
%pylab inline 
from turtle import *
```

```python
#Ask for the number of the Koch snowflake 
stp = int(input("Please enter the number of step of Koch snowflake that you want:"))

#Set the initial shape of the triangle
#The letters F=forward (move), L=left (turn) and R=Right (turn) are the instructions for the 'turtle'
base = "FLFLF"
#According to the number of step change the initial shape for the new shape's instructions  
for i in range(stp):
    base = base.replace("F","FRFLFRF")

#Customize the scren that show the image
turtle.setup(300, 300)
turtle.speed(100)
turtle.clear()
turtle.title('Koch snowflake')
turtle.hideturtle()
turtle.penup() # Move the turtle to the initial coordinates
turtle.setx(-70)
turtle.sety(-50)
turtle.pendown()

#Executes the instructions save in base
for move in base:
    if move == "F":
        turtle.forward(50.0 / (3 ** (stp - 1))) #Important to recalibrate the scale!
    elif move == "L":
        turtle.left(120)
    elif move == "R":
        turtle.right(60)

turtle.exitonclick()

        
```
