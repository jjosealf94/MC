```python
%pylab inline
from scipy.fftpack import fftshift as fts
from scipy import interpolate
from scipy.optimize import fminbound
```

```python
#Crea una matriz de ceros e inicializa con unos un rectángulo de lado 32 en el centro
matriz=zeros((1024,1024))
for i in range(496,528,1):
    for j in range(496,528,1):
        matriz[i][j]=1

#Grafica el resultado obtenido
figure(figsize(5,5))
imshow(matriz)
colorbar()
        
```

```python
#Aplica Fourier al arreglo con el rectángulo encontrado y realiza un corrimiento
DFmatriz=fft2(matriz)
SDFmatriz=fftshift(DFmatriz)

#Crea un nuevo arreglo para asignar un cuadrado de lado 200 producto de la DFT corrida
centro=zeros((200,200))
# Indices k y m para matriz centro [200x200]
k=0
m=0
#Indices i y j para matriz con la DFT corrida
for i in range(412,612,1):
    for j in range(412,612,1):
            centro[k][m]= log((abs(SDFmatriz[i][j])**2)+1) # Se suma 1 para evitar warning, log resalta
            m=m+1
            if m>199:
                m=0
    k+=1

#Grafica el resultado obtenido
fig=figure(figsize(10,5))
ax1 =subplot2grid((1,2),(0,0))
ax2 =subplot2grid((1,2),(0,1))
ax1.imshow(matriz,cmap='gray')
ax2.imshow(centro)
ax2.set_title('Patrón de interferencia de una apertura cuadrada \n (log resaltado)')
```

```python
#Genera la matriz de ceros e inicializa la doble rendija con valores de unos
#Altura 200, Grosor 20
#Separadas 104 pixeles una de otra
matriz=zeros((1024,1024))
for i in range(412,612,1):
    for j in range(450,470):
        matriz[i][j]=1
        matriz[i][j+104]=1        
```

```python
#Aplica Fourier al arreglo con la doble rendija y realiza un corrimiento
DFmatriz=fft2(matriz)
SDFmatriz=fftshift(DFmatriz)

#Crea un nuevo arreglo para asignar un cuadradao de lado 128 producto del arreglo con DFT corrida
centro=zeros((128,128))
#Indices k y m para nuevo arreglo
k=0
m=0
#Indices i y j para arreglo con la DFT corrida
for i in range(448,576,1):
    for j in range(448,576,1):
            centro[k][m]=(abs(SDFmatriz[i][j])**2) #Se toma sólamente la norma
            m=m+1
            if m>127:
                m=0
    k+=1

#Se genera la gráfica del resultado
fig=figure(figsize(10,5))
ax1 =subplot2grid((1,2),(0,0))
ax2 =subplot2grid((1,2),(0,1))
ax1.imshow(matriz,cmap='gray')
ax2.imshow(centro)
ax2.set_title('Patrón de interferencia de una doble rendija')
```

```python
#Crea la matriz de ceros e inicializa un circulo de radio 0.01*1024 dentro con valores de unos
matriz=zeros((1024,1024))
radio=0.01
r=radio*1024
for i in range(1024):
    for j in range(1024):
        d=(i-512)**2+(j-512)**2
        if d<r**2:
            matriz[i][j]=1
            
#Se aplica DFT al arreglo obtenido y se le aplica un corrimiento
DFmatriz=fft2(matriz)
SDFmatriz=fftshift(DFmatriz)

#Se genera la matriz que contiene el patrón de airy el cúal va ser el arreglo con 
# la norma del arreglo con la DFT corrida 
airy=zeros((1024,1024))
#Indices k y m para la matriz con el patrón
k=0
m=0
for i in range(1024):
    for j in range(1024):
            airy[k][m]=(abs(SDFmatriz[i][j])**2) #Sólo la norma, luego se resalta con log
            m=m+1
            if m>1023:
                m=0
    k+=1

#Se define las abcisas comparables
abcisas=2*pi*(0.01)/1.*linspace(-512,512,1024)
#Se encuentra la intensidad cero y se crea un vector que pase por el centro con las intensidades
Io=max(airy[512])
I=(airy[512])
#Se normaliza el arreglo de intensidades
Inor=I/Io

#Se genera la figura
fig=figure(figsize(10,8))

#Se crean los ejes para las tres figuras y se ubican de acuerdo a lo sugerido
ax1 =subplot2grid((3,2),(0,0),rowspan=1,colspan=2)
ax2 =subplot2grid((3,2),(1,0),rowspan=2)
ax3 =subplot2grid((3,2),(1,1),rowspan=2)

#Se grafica la intensidad normalizada contra las abcisas comparables
ax1.plot(abcisas,Inor,'-d')
ax1.set_xlim([0,10])
ax1.set_ylim([0,1])
ax1.set_xlabel('$2\pi Rj/L$')
ax1.set_ylabel('$I/I_0$')

#Se grafica el arreglo creado que representa la apertura circular
ax2.imshow(matriz,cmap='gray')
#Se resalta con logaritmo (+1 evita warning) el arreglo con DFT corrida y se grafica
airy =log(airy+1)
ax3.imshow(airy)
ax3.set_title('Patrón de interferencia de una apertura circular (Patrón de Airy)\n log resaltado')
fig.tight_layout()
show()

```

```python
#Se crean los arreglos correspondientes a mínimos y máximos (X y Y) sugeridos en la tabla
xminTeo=[3.833,7.016,10.174]
xmaxTeo=[0,5.136,8.417,11.620]
yminTeo=[0.,0.,0.]
ymaxTeo=[1.,0.0175,0.0042,0.0016]

#Se crean arreglos para guardar mínimos y máxmios (X y Y) producto de la interpolación
xminInter=[0,0,0]
xmaxInter=[0,0,0,0]
yminInter=[0,0,0]
ymaxInter=[0,0,0,0]

#Se realiza una interpolación cúbica 
interpol_cubic = interpolate.interp1d(abcisas,Inor,kind='cubic')
#Se define la función negativo de la interpolación (para calcular máximos)
def nInterpol_cubic(x):
    return -1*interpol_cubic(x)

#Se definen los rangos en x para el mínimo que límitarán la busqueda de los mínimos del polinomio 
#producto de la interpolación
xlowminRan=[(xminTeo[0]-0.5),(xminTeo[1]-0.5),(xminTeo[2]-0.5)]
xupminRan=[(xminTeo[0]+0.5),(xminTeo[1]+0.5),(xminTeo[2]+0.5)]
#Se definen los rangos en x para el máximo que límitarán la busqueda de los máximos del polinomio 
#de la interpolación
xlowmaxRan=[(xmaxTeo[0]-0.5),(xmaxTeo[1]-0.5),(xmaxTeo[2]-0.5),(xmaxTeo[3]-0.5)]
xupmaxRan=[(xmaxTeo[0]+0.5),(xmaxTeo[1]+0.5),(xmaxTeo[2]+0.5),(xmaxTeo[3]+0.5)]

#Se buscan los mínimos en los rangos para el polinomio interpolación
for i in range(len(xminTeo)):
    mini = fminbound(interpol_cubic,xlowminRan[i],xupminRan[i],full_output=1)
    xminInter[i]=mini[0]
    yminInter[i]=interpol_cubic(xminInter[i])
#Se buscan los MÍNIMOS en  los rangos para el polinomio negativo de la interpolación
for i in range(len(xmaxTeo)):
    maxi = fminbound(nInterpol_cubic,xlowmaxRan[i],xupmaxRan[i],full_output=1)
    xmaxInter[i]=maxi[0]
    ymaxInter[i]=interpol_cubic(xmaxInter[i])
    

#Se grafican los resultados obtenidos explicación leyendas
tlin=linspace(0,12,100)
figure(figsize(12,3))
plot(tlin,interpol_cubic(tlin),color='c',label='cubic fit')
scatter(xminTeo,yminTeo,color='b',label='mínimos teóricos')
scatter(xmaxTeo,ymaxTeo,color='g',label='máximos teóricos')
scatter(xminInter,yminInter,color='y',label='mínimos interpolados')
scatter(xmaxInter,ymaxInter,color='r',label='máximos interpolados')
grid()
title('Comparativa de mínimos y máximos teóricos e interpolados')
legend()
xlim(0,13)
show()
```

```python
import numpy as np

#Se definen arreglos para el error en X y Y de los mínimos y máximos interpolados
xminerror=[0,0,0]
xmaxerror=[0,0,0,0]
yminerror=[0,0,0]
ymaxerror=[0,0,0,0]

#Se utilizan sólamente 4 cifras de decimales ya que en los valores teóricos se tienen máximo 4
#cifras decimales
#Se realiza el error porcentual de los máximos y los mínimos
for i in range(3):
    xminerror[i]=(abs(np.round(xminInter[i],4)-xminTeo[i])/xminInter[i])*100
    yminerror[i]=(abs(np.round(yminInter[i],4)-yminTeo[i])/yminInter[i])*100

for i in range(4):
    xmaxerror[i]=(abs(np.round(xmaxInter[i],4)-xmaxTeo[i])/xmaxInter[i])*100
    ymaxerror[i]=(abs(np.round(ymaxInter[i],4)-ymaxTeo[i])/ymaxInter[i])*100

#Se imprime la información de los mínimos y los máximos por separada
print("xminTeo xminInter %serror yminTeo yminInter %serror\n"%('%','%'))
for i in range(3):
    print ("%6.3f  %6.3f  %6.3f%s  %6.3f  %6.3f  %6.3f%s\n"%(xminTeo[i],xminInter[i],xminerror[i],'%',yminTeo[i],yminInter[i],yminerror[i],'%'))

print("xmaxTeo xmaxInter %serror ymaxTeo ymaxInter %serror\n"%('%','%'))
for i in range(4):
    print ("%6.3f  %6.3f  %6.3f%s  %6.3f  %6.3f  %6.3f%s\n"%(xmaxTeo[i],xmaxInter[i],xmaxerror[i],'%',ymaxTeo[i],ymaxInter[i],ymaxerror[i],'%'))

```

En el caso del primer máximo al ser este cero, al calcularse el error se tiene que cualquier número restado cero es él mismo y luego divido él mismo da 1, que al multiplicarse por 100 muestra un error del 100%.
