```python
#Lagrange Function
def lagrange(N,flag=False):
    L=0. #Return the lagrange polynomial 
    t=var('t')
    #Do the sum
    for j in range(N):
        lj=1 #Lagrange Base
        tj = var('t%d'%j)
        #Do the product
        for i in range(N):
             if (j != i):
                ti = var('t%d'%i)
                lj *= (t - ti)/(tj - ti) #Compute the lagrange base
        L+=var('f%d'%j)*lj #Compute the lagrange polynomials
    return(L)
```

```python
lagrange(2)
```

```python
def adams(N):
    #Lagrange Function with ti and tj  modification
    L=0.
    t=var('t')
    h=var('h')
    for j in range(0,N,1):
        lj=1
        tj = var('t0') - (j*h)
        for i in range(0,N,1):
            if (j != i):
                ti = var('t0') - (i*h)
                lj *= (t - ti)/(tj - ti)
        L+=var('f%d'%j)*lj
    adam=simplify(integrate(L,(t,t0,t0+h))) #Integrate the the lagrange polynomial and simplifies
    return (adam) 
```

#Ecuación de Poisson para campo gravitacional

```python
%pylab inline
```

```python
G = 6.673*10**(-11.) #Constante de Gravitación universal
p=400. #Densidad del cubo en Kg/m3
L=3000. #Lado del espacio en m
l=270. #Lado del cubo maciso en m
h=30. #Separación entre los puntos en m

N=L//h #Numero de puntos del espacio
if N%2==0:
    N=N+1
N=int(N)

nCubo=l//h #Numero de puntos del cubo
if nCubo%2==0:
    nCubo=nCubo+1
nCubo=int(nCubo)

centro=(N+1)/2 #Toma las coordenadas del centro del espacio

#Establece la discretización del espacio
coords=np.linspace(-L/2,L/2,N)


#Vectoriza las coordenadas
xmesh,ymesh,zmesh=np.meshgrid(coords,coords,coords)


r=sqrt(xmesh**2 +ymesh**2 + zmesh**2)+1 #Toma la norma del vector posición 
Mtot=p*(l**3) #La masa total densidad*Volumen
A=Mtot*G/1500. #Constante que establece a 0 el potencial en la frontera

potencial=-G*Mtot/r + A #Potencial de una masa puntual 

#Condiciones de Dirichlet potencial 0 en la frontera
potencial[0] = 0.
potencial[-1] = 0.
potencial[:,0] = 0.
potencial[:,-1] = 0.
potencial[:,:,0] = 0.
potencial[:,:,-1] = 0.

#Estabelce la coordenadas de inicio y fin de cada lado del cubo 
ini=int(floor(centro-(nCubo/2)))
fin=int(floor(centro+(nCubo/2)))

          
```

```python
numiter = 1000
#Realiza la estimación para la ecuación de Poisson por medio del método de la relajación
for __ in range(numiter):
     
    potencial=(1/6.)* (  roll(potencial,1,axis=0) + roll(potencial,-1,axis=0) 
                       + roll(potencial,1,axis=1) + roll(potencial,-1,axis=1)
                       + roll(potencial,1,axis=2) + roll(potencial,-1,axis=2) )
    
    potencial[ini:fin,ini:fin,ini:fin] -= (1./6.)*4.*pi*G*p*h**2. #Se especifica el valor del cubo
    #En cada interación se debe establecer nuevamente las condiciones de Dirichlet
    potencial[0] = 0.
    potencial[-1] = 0.
    potencial[:,0] = 0.
    potencial[:,-1] = 0.
    potencial[:,:,0] = 0.
    potencial[:,:,-1] = 0. 
```

```python
figure(figsize=(10,10))
pcolor(coords,coords,potencial[centro]*10000,cmap='gist_heat') #Se multiplica por 10000 para establecer unidades
axis('scaled')
colorbar()
#Establece la líneas equipotenciales 
clabel(contour(coords,coords,potencial[centro]*10000,levels=[-1,-2,-4,-8,-16],linestyles='dashed',colors='k') )
xlabel('$y/m$',fontsize=15)
ylabel('$z/m$',fontsize=15)
title('$\phi/cm^2s^{-1}$',fontsize=20)
show()
```

```python
#Crea un nuevo arreglo para establecer el potencial luego de segmentar el cubo
potencialmp = np.zeros((N,N,N)) 
#Establece las coordenadas de los cubos más pequeños
m = linspace(-120.,120.,9)

#Se suma el potencial de cada masa puntual
for i in range(len(m)):
    for j in range(len(m)):
        for k in range(len(m)):
            #Se establece la norma del vector posición para cada masa puntual
            rm=sqrt((xmesh-m[i])**2.+(ymesh-m[j])**2.+(zmesh-m[k])**2.)+3.
            potencialmp += (-G*Mtot/729.)/rm
```

```python
figure(figsize=(10,6))
#Muestra la comparación de los métodos
plot(coords,potencial[50,50,:]*10000, label = 'sum $+\epsilon$') 
scatter(coords,(potencialmp[50,50,:]+A)*10000,facecolor='none',label = 'Poisson') 
xlabel('$z/m$',fontsize=20)
ylabel('$\phi/cm^2s^{-1}$',fontsize=20)
title('Potencial Gravitacional',fontsize=20)
legend()
show()
```

```python
#Se calcula el campo como la derivada del potencial
campo=(roll(potencial[50,50,:],-1)-roll(potencial[50,50,:],1))/(2.*h) 
campo = campo[1:-1] #Sin significado físico

figure(figsize=(10,6))
scatter(coords[1:-1],-campo*1000000,facecolors='none') #Se coloca en unidades adecuadas
xlabel('$z/m$',fontsize=15)
ylabel('$g/\mu m^2s^{-1}$',fontsize=15)
title('Campo Gravitacional',fontsize=20)
show()
```

```python
from scipy import interpolate

#Se interpola el campo en uno de los ejes 
campoFun = interpolate.interp1d(coords[1:-1],-campo,kind='cubic')

#Se describe función que define la ecuación diferencial a resolver, donde Y(z,vz)
def f(Y):
    return np.array([Y[1],campoFun(Y[0])])

#RK4
def one_Kutta4_step():
    global t
    global Deltat
    global Y
    K1 = f(Y)
    K2 = f(Y + Deltat*(1./2.*K1))
    K3 = f(Y + Deltat*(1./2.*K2))
    K4 = f(Y + Deltat*K3)
    Y+= Deltat * (1./6.*K1 + 1./3.*K2 + 1./3.*K3 + 1./6.*K4)
    t+= Deltat

#Se tomaron estos pasos para que la masa llegara 0, no necesariamente la cara 
numIter = 25085
#Se establece z0 teniendo en cuenta que la mitad del cubo ocupa 135m  
#Además se quiere tener a una distancia de 500m a partir de la cara
z0= 635.
#Se establecen las condiciones inciales
Y = np.array([z0,0.])
t = 0. 
Deltat = 1.
history = np.array([[t,Y[0],Y[1]]]) 
#Se aplica RK4 durante el # de iteraciones
for __ in range(numIter):
    one_Kutta4_step()
    history = np.append(history,[[t, Y[0], Y[1]]],axis=0)

#Se extrae la información del tiempo y la altura
t = history[:,0]
z = history[:,1]



```

```python
#Variables que guardan la posición del cubo, el tiempo en que colisionan y
#el indice en el arreglo al que corresponden
zcubo=0.
tcubo=0.
ind=0.
#Función que busca una altura aproximada a 135m con un epsilon de 0.05
for i in range(len(z)-1):
               if(z[i]<135.05 and z[i]>134.95):
                    tcubo=t[i]
                    zcubo=z[i]
                    ind=i

#Se muestra la información
figure(figsize=(7,5))
plot(t[:ind],z[:ind])
plot(tcubo,zcubo-20,'s',markersize=20)
xlabel(r'$t/seg.$',fontsize=20)
ylabel(r'$z/m$',fontsize=20)
grid()
show()

print('Colisiona en:',round(tcubo/3600.,3),'horas')
```
